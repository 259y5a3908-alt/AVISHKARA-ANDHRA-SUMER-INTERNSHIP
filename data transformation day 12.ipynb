```python
import pandas as pd
import numpy as np

# Load CSV file
df = pd.read_csv("sales.csv")

print("Original Data:")
print(df)

# ----------------------------
# DATA TRANSFORMATION
# ----------------------------

# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Convert customer names to uppercase
df['customer_name'] = df['customer_name'].str.upper()

# ----------------------------
# FEATURE CREATION
# ----------------------------

# Total Amount
df['total_amount'] = df['quantity'] * df['price']

# Extract Order Month
df['order_month'] = df['order_date'].dt.month

# Extract Order Year
df['order_year'] = df['order_date'].dt.year

# High Value Order Flag
df['high_value_order'] = np.where(df['total_amount'] >= 10000, 1, 0)

# Product Category
df['category'] = np.where(
    df['price'] >= 10000,
    'Premium',
    'Regular'
)

print("\nTransformed Data:")
print(df)

# Save transformed dataset
df.to_csv("sales_transformed.csv", index=False)
```
