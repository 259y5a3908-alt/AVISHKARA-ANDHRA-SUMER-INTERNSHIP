```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("products-100.csv")
numeric_df = df.select_dtypes(include=['int64', 'float64'])
print("Numeric Columns:")
print(numeric_df.columns)
# 1. Line Plot
if len(numeric_df.columns) >= 1:
    plt.figure(figsize=(6,4))
    plt.plot(numeric_df.iloc[:,0])
    plt.title("Line Plot")
    plt.xlabel("Index")
    plt.ylabel(numeric_df.columns[0])
    plt.show()

# 2. Bar Plot
if len(numeric_df.columns) >= 1:
    plt.figure(figsize=(6,4))
    numeric_df.iloc[:,0].head(10).plot(kind='bar')
    plt.title("Bar Plot")
    plt.xlabel("Index")
    plt.ylabel(numeric_df.columns[0])
    plt.show()

# 3. Histogram
if len(numeric_df.columns) >= 1:
    plt.figure(figsize=(6,4))
    plt.hist(numeric_df.iloc[:,0].dropna(), bins=10)
    plt.title("Histogram")
    plt.xlabel(numeric_df.columns[0])
    plt.ylabel("Frequency")
    plt.show()

# 4. Box Plot
if len(numeric_df.columns) >= 1:
    plt.figure(figsize=(6,4))
    plt.boxplot(numeric_df.iloc[:,0].dropna())
    plt.title("Box Plot")
    plt.show()

# 5. Heatmap (without seaborn)
if len(numeric_df.columns) > 1:
    corr = numeric_df.corr()

    plt.figure(figsize=(6,5))
    plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
    plt.colorbar()
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.title("Correlation Heatmap")
    plt.tight_layout()
    plt.show()
else:
    print("Need at least 2 numeric columns for Heatmap.")
```
