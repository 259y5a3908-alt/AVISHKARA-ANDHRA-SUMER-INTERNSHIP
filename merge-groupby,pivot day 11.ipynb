```python
import pandas as pd

employees = pd.read_csv("employee.csv")
departments = pd.read_csv("department.csv")

merged_data = pd.merge(
    employees,
    departments,
    on="Department",
    how="inner"
)

print(merged_data)

avg_salary = merged_data.groupby("Department")["Salary"].mean()
print(avg_salary)

report = merged_data.groupby("Department").agg(
    Employee_Count=("EmpID", "count"),
    Average_Salary=("Salary", "mean"),
    Maximum_Salary=("Salary", "max"),
    Minimum_Salary=("Salary", "min")
)
print(report)

high_salary = merged_data[merged_data["Salary"] > 60000]
print(high_salary)

sorted_data = merged_data.sort_values(
    by="Salary",
    ascending=False
)
print(sorted_data)

merged_data["Bonus"] = merged_data["Salary"] * 0.10
print(
    merged_data[
        ["EmpID", "Name", "Salary", "Bonus"]
    ]
)

pivot = pd.pivot_table(
    merged_data,
    values="Salary",
    index="Department",
    aggfunc=["mean", "max", "min"]
)
print(pivot)

top5 = merged_data.nlargest(
    5,
    "Salary"
)
print(top5)

dept_summary = merged_data.groupby(
    "Department"
).agg(
    Total_Employees=("EmpID", "count"),
    Average_Salary=("Salary", "mean"),
    Manager=("Manager", "first"),
    Location=("Location", "first"),
    Budget=("Budget", "first")
)
print(dept_summary)

merged_data.to_csv(
    "processed_employee_data.csv",
    index=False
)

print("Processed file saved successfully.")
```
