```python


import pandas as pd
import numpy as np
# Creating dataset
data = {
    'Name': ['Ali', 'Sara', 'John', 'Sara', np.nan],
    'Age': [22, 25, np.nan, 25, 30],
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Mumbai', np.nan]
}
df = pd.DataFrame(data)
# Display Original Data
print("Original Dataset:\n")
print(df)
# MISSING VALUES
print("\nMissing Values Count:\n")
print(df.isnull().sum())
# Filling missing values
df['Name'] = df['Name'].fillna("Unknown")
# Mean is used because Age is numerical
age_mean = df['Age'].mean()
df['Age'] = df['Age'].fillna(age_mean)
# Mode is used because City is categorical
city_mode = df['City'].mode()[0]
df['City'] = df['City'].fillna(city_mode)
print("\nDataset After Filling Missing Values:\n")
print(df)
print("\nData Types:\n")
print(df.dtypes)
# Convert Age float to integer
df['Age'] = df['Age'].astype(int)
print("\nData Types After Conversion:\n")
print(df.dtypes)

# -------------------------------
# DUPLICATE VALUES
# -------------------------------

print("\nDuplicate Rows:\n")
print(df.duplicated())

# Remove duplicate rows
df = df.drop_duplicates()

print("\nDataset After Removing Duplicates:\n")
print(df)
```
